```python
import pandas as pd

df = pd.read_csv('IT_vacancies.csv', delimiter=',', encoding='utf-8', quotechar='"')
df.info()
```

```python
df['key_skills'] = df['key_skills'].fillna("{}")
df['key_skills'] = df['key_skills'].apply(lambda x: set(map(lambda y: y.strip().lower() ,x.replace('"', '').strip('{}').split(','))))
```

```python
df['key_skills']
```

```python
skills_set = set()
for skills in df['key_skills']:
    skills_set.update(skills)
```

```python
len(skills_set)
```

```python
df2 = pd.read_csv('IT_vacancies_full.csv', delimiter=',', encoding='utf-8', quotechar='"')
df2.info()
```

```python
df2['Keys'] = df2['Keys'].apply(lambda x: set(map(lambda y: y.strip().lower() ,x.replace("'", '').strip('[]').split(','))))
df2['Keys']
```

```python
for skills in df2['Keys']:
    skills_set.update(skills)
```

```python
skills_set
```

```python
new_df = df[['id', 'name', 'key_skills']].rename(
    columns={
        'key_skills': 'skills'
    }
)
new_df = pd.concat([new_df, df2[['Ids', 'Name', 'Keys']].rename(
    columns={
        'Ids': 'id',
        'Name': 'name',
        'Keys': 'skills',
    }
)], ignore_index=True)
new_df.info()
```

```python
new_df.to_csv('IT_vacancies_for_prof_map.csv', index=False)
```
